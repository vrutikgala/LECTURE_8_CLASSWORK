# Functions

```python
# Setup
import pandas as pd
```

```python
# score processor function
def get_scores(report_card_dictionary,division_name):
    
    # Convert the dictionary into a dataframe
    report_card_df=pd.DataFrame(report_card_dictionary)
    
    # Add a column called division
    
    report_card_df["division"]=division_name
    
    # Add a column called total score
    
    report_card_df["total_score"]= report_card_df["math_scores"]+ report_card_df["chemistry_scores"]+report_card_df["information_technology_scores"]+report_card_df["physics_scores"]
    
    # Add a column called average score
    
    report_card_df["average_score"]=report_card_df["total_score"]/4
    
    # Add a column called status
    
    report_card_df.loc[report_card_df["average_score"]>=35 , "status"]= "Pass"
    report_card_df.loc[report_card_df["average_score"]<35  , "status"]="Fail"
    
    
    # add a colum called character grades
    
    report_card_df.loc[report_card_df["average_score"]>=90 , "character_grade"]="A"
    report_card_df.loc[(report_card_df["average_score"]>=75)&(report_card_df["average_score"]<90),"character_grade"]="B"
    report_card_df.loc[(report_card_df["average_score"]>=50)&(report_card_df["average_score"]<75),"character_grade"]="C"
    report_card_df.loc[(report_card_df["average_score"]>=35)&(report_card_df["average_score"]<50),"character_grade"]="D"
    report_card_df.loc[report_card_df["average_score"]<35 , "character_grade"]="F"
    
    
    
    
    
    
    return report_card_df
    
    
```

## Scores and email function

```python
def get_scores_and_emails_combined(report_card_df,emails_df):
    scores_and_email_df=pd.merge(report_card_df,emails_df,how = "left",on=["student_name","division"])
    return scores_and_email_df
```

```python
def get_all_division_scoresheet(div_a_df,div_b_df,div_c_df):
    all_division_df=pd.concat([div_a_df,div_b_df,div_c_df],ignore_index= True)
    return all_division_df
```
